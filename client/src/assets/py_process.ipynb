```python
import pandas as pd
import json
from iso639 import Lang

with open("./json/countries.json", "r", encoding='utf-8') as file:
  fileStr = file.read()
with open("./json/num_speakers.json", "r", encoding='utf-8') as file:
  fileStr2 = file.read()

arr = json.loads(fileStr)
arr2 = json.loads(fileStr2)


df = pd.DataFrame(arr)
df2 = pd.DataFrame(arr2)


df['languagename'] = df['language'].apply(lambda x: x['name'])
df['languagecode'] = df['language'].apply(
    lambda x: x['code']).astype(str).apply(lambda x: x.lower())
df = df.rename({'isoCode': 'isocode'}, axis=1)


# df.loc[df['languagecode'] == "bn", 'speakers']
df
```

```python

df['code']=df['code'].apply(lambda x:x.lower())
df['region'] = df['region'].apply(lambda x: x.lower())
df = df[['name', 'code', 'region', 'languagename', 'languagecode', 'isocode']]
df=df.sort_values('languagename')
df = df.reset_index(drop=True)


for i in range(len(df2)):
  df.loc[df['code'] == df2['countrycode'][i], 'speakers'] = df2['speakers'][i]

# fill nan values with 0
df['speakers'] = df['speakers'].fillna(0)


df


```

```python
def safe_pt3(x):
    try:
        return Lang(x).pt3
    except Exception as e:
        return 'Unknown'


df['iso6393'] = df['languagecode'].apply(lambda x: safe_pt3(x))
def safe_pt1(x):
    try:
        return Lang(x).pt1
    except Exception as e:
        return 'Unknown'


df['iso6391'] = df['languagecode'].apply(lambda x: safe_pt1(x))


df
```

```python
langs = {}
df = df.sort_values('speakers', ascending=False)
df = df.reset_index(drop=True)
for i in range(0, len(df)):
    if df['languagecode'][i] in langs:
        langs[df['languagecode'][i]]['countrynames'].append(df['name'][i])
        langs[df['languagecode'][i]]['countrycodes'].append(df['code'][i])
        langs[df['languagecode'][i]]['speakers'].append(df['speakers'][i])
    else:
        langs[df['languagecode'][i]] = {
                                'languagecode':df['languagecode'][i],
                                'iso6393':df['iso6393'][i],
                                'iso6391':df['iso6391'][i],
                                'languagename':df['languagename'][i],
                                'countrynames':[df['name'][i]],
                                'countrycodes':[df['code'][i]], 
                                'isocode':df['isocode'][i],
                                'speakers':[df['speakers'][i]]
                                  }
    
langs
    
```

```python

with open("./json/langs_out.json", "w", encoding='utf-8') as file:
    file.write(json.dumps(langs, ensure_ascii=False))
```
